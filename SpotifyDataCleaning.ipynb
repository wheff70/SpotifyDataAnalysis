```python
import pandas as pd
```

```python
spotify_data = pd.read_csv("spotify_top_songs_audio_features.csv")
spotify_data.head()
```

```python
spotify_data.count
```

#### Removing irrelevant attributes of dataset

```python
spotify_data.drop(columns= "duration_ms", inplace=True)
spotify_data.head()
```

```python
spotify_data.count
```

#### Dropping irrelevant rows

```python
spotify_data.drop_duplicates(inplace=True)
spotify_data.dropna(inplace=True)
spotify_data.count
```

#### Saving updated dataset to .csv

```python
spotify_data.to_csv("SpotifyData.csv", index=False)
```
